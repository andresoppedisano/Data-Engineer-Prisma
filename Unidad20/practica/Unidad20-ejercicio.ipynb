```python
# 1. Conectarse a la url que contiene el archivo CSV de medallas olímpicas.


import pandas as pd
import sqlite3

medals=pd.read_csv('C:/Data-Engineer-Prisma/Unidad20/contenido/medals.csv',encoding = "utf-8")
```

```python
# 2. Descargar los datos y obtener un sub dataset que contenga a todas las 
# medallas de oro (Gold) Estados Unidos (USA) a partir delaño 1950.


usa_champions = medals[(medals["Year"] > 1950) & (medals["Medal"] == "Gold") & (medals["NOC"] == "USA")]
usa_champions
```

```python
# 3. Crear una base de datos “olympics” en SQLite y la tabla “medals"


conn = sqlite3.connect('olympics.db')

cursor = conn.cursor()

query = """Create Table If Not Exists medals(
                                    Year Integer,
                                    City Text,
                                    Sport Text,
                                    Discipline Text,
                                    NOC Text,
                                    Event Text,
                                    Event_gender Text,
                                    Medal Text
                                    );"""

cursor.execute(query)
conn.commit()
```

```python
# 4. Guardar en la base de datos. los datos generados en el sub dataset.


data = list(usa_champions.itertuples(index=False, name=None))
cursor.executemany("Insert Into medals Values (?, ?, ?, ?, ?, ?, ?, ?)", data)
conn.commit()
```

```python
# 5. Consultar la base de datos y validar que los datos se hayan cargado correctamente.


query = "Select * From medals Limit 20"
cursor.execute(query)
results = cursor.fetchall()

for r in results:
    print(r)
    
conn.close()
```
